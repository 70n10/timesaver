```python
import requests
from bs4 import BeautifulSoup
import time
def classement(link):
    page = requests.get(link)

    soup = BeautifulSoup(page.content, 'html.parser')
    
    col1 = []
    for i in soup.find_all(class_="ranCrang"):
        col1.append(i.get_text())
        
    col2 = []
    for i in soup.find_all(class_="ranCteam"):
        col2.append(i.get_text())

    col3 = []
    for i in soup.find_all(class_="ranCsp"):
        col3.append(i.get_text())

    col4 = []
    for i in soup.find_all(class_="ranCs"):
        col4.append(i.get_text())

    col5 = []
    for i in soup.find_all(class_="ranCu"):
        col5.append(i.get_text())

    col6 = []
    for i in soup.find_all(class_="ranCn"):
        col6.append(i.get_text())

    col7 = []
    for i in soup.find_all(class_="ranCstrp"):
        col7.append(i.get_text())

    col8 = []
    for i in soup.find_all(class_="ranCtg"):
        col8.append(i.get_text())

    col9 = []
    for i in soup.find_all(class_="ranCte"):
        col9.append(i.get_text())

    col10 = []
    for i in soup.find_all(class_="ranCpt"):
        col10.append(i.get_text())

    res = ''
    #print(soup.find(id="ctl01_ctl14_MatchcenterMaster_ctl01_pHeader").get_text())
    for i1, i2, i3, i4, i5, i6, i7 ,i8, i9, i10 in zip(col1, col2, col3, col4, col5, col6, col7, col8, col9, col10):
        res += '\t' + i1 +'\t' +  i2.replace('FC','').strip()+'\t'+  i3+'\t'+ i4+'\t'+ i5+'\t'+ i6+'\t'+ i7+'\t'+ i8+'-'+i9+'\t'+ i10+'\n'
    return res


def matches(link):
    page = requests.get(link)

    soup = BeautifulSoup(page.content, 'html.parser')
    def testjoue():
        hidden = []
        for i in soup.find_all(class_="col-md-2 goaldivider hidden-xs hidden-sm"):
            hidden.append(i.get_text().strip())

        joueOuNon = []
        for i in range(len(hidden)):
            try:
                if hidden[i] == '-' and hidden[i+1]==':':
                    joueOuNon.append(True)
                elif hidden[i] == ':':
                    pass
                else:
                    joueOuNon.append(False)
            except:
                joueOuNon.append(False)
        return joueOuNon

    team1 = []
    for i in soup.find_all(class_="col-md-5 col-xs-12 teamA"):
        team1.append(i.get_text().strip())

    team2 = []
    for i in soup.find_all(class_="col-md-5 col-xs-12 teamB"):
        team2.append(i.get_text().strip())

    score1 = []
    for i in soup.find_all(class_="col-xs-12 col-md-5 torA"):
        score1.append(i.get_text().strip())

    score2 = []
    for i in soup.find_all(class_="col-xs-12 col-md-5 torB"):
        score2.append(i.get_text().strip())
    res = ''
    i=0
    for t1, t2, indic in zip(team1, team2, testjoue()):
        try:
            if indic == True:
                res += t1.replace('FC', '').strip() + ' - ' + t2.replace('FC','').strip() +'\t' + score1[i] +'-' + score2[i] +'\n'
                i+=1
            else:
                res += t1.replace('FC', '').strip() + ' - ' + t2.replace('FC', '').strip() + '\t' +  'X-X' +'\n'

        except:
            res+=('erreur')
    return res

```
